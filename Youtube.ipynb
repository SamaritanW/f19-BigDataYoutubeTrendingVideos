# Data Cleaning

```python
#Check how many NaNs are in each column
df_youtubeUS.isnull().sum()
```

```python
#Replace NaNs with empty string
df_youtubeUS.fillna('',inplace = True)
df_youtubeUS.isnull().sum()
```

```python
df_youtubeCA.fillna('',inplace = True)
df_youtubeCA.isnull().sum()
```

```python
df_youtubeGB.fillna('',inplace = True)
df_youtubeGB.isnull().sum()
```
